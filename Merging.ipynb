```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import os
from IPython.display import clear_output
from sympy import Polygon, Point
from tqdm import tqdm
import math
```

Merging trave points

```python
trave1 = pd.read_csv('C:/GNG2/travepoint/ku2022_11_travpoint/ku2022_11_trav_point_grid3.csv') 
trave2 = pd.read_csv('C:/GNG2/travepoint/ku2022_12_travpoint/ku2022_12_trav_point_grid3.csv')
trave3 = pd.read_csv('C:/GNG2/travepoint/ku2022_13_travpoint/ku2022_13_trav_point_grid3.csv')
trave4 = pd.read_csv('C:/GNG2/travepoint/ku2022_14_travpoint/ku2022_14_trav_point_grid3.csv')
trave5 = pd.read_csv('C:/GNG2/travepoint/ku2022_15_travpoint/ku2022_15_trav_point_grid3.csv')
trave6 = pd.read_csv('C:/GNG2/travepoint/ku2022_16_travpoint/ku2022_16_trav_point_grid3.csv')

trave = pd.concat([trave1,trave2,trave3,trave4,trave5,trave6])
trave = trave.iloc[:,[1,2]].to_numpy()
plt.scatter(*trave.T, s=1, c='green',label='Traversable') 
```

Merging location data

```python
loc1 = pd.read_csv('C:\ku2022_1\ku2022_11_locopt.csv') 
loc2 = pd.read_csv('C:\ku2022_1\ku2022_12_locopt.csv')
loc3 = pd.read_csv('C:\ku2022_1\ku2022_13_locopt.csv')
loc4 = pd.read_csv('C:\ku2022_1\ku2022_14_locopt.csv')
loc5 = pd.read_csv('C:\ku2022_1\ku2022_15_locopt.csv')
loc6 = pd.read_csv('C:\ku2022_1\ku2022_16_locopt.csv')

loc = pd.concat([loc1,loc2, loc3,loc4, loc5, loc6])
loc = loc.iloc[:,[2,3]].to_numpy()
plt.scatter(*loc.T, s=1, c='blue',label='Location') 
```

Merging prox data

```python
prox1 = pd.read_csv('C:\ku2022_1\ku2022_11_absproxopt.csv') 
prox2 = pd.read_csv('C:\ku2022_1\ku2022_12_absproxopt.csv')
prox3 = pd.read_csv('C:\ku2022_1\ku2022_13_absproxopt.csv')
prox4 = pd.read_csv('C:\ku2022_1\ku2022_14_absproxopt.csv')
prox5 = pd.read_csv('C:\ku2022_1\ku2022_15_absproxopt.csv')
prox6 = pd.read_csv('C:\ku2022_1\ku2022_16_absproxopt.csv')

prox = pd.concat([prox1,prox2, prox3, prox4, prox5, prox6])
prox = prox.iloc[:,[2,3]].to_numpy()
plt.scatter(*prox.T, s=1, c='red',label='Proximity')
plt.xticks(np.arange(-40, 130, 40))
plt.yticks(np.arange(-180, 30, 40))
```

```python
plt.scatter(*trave.T, s=0.5, c='green', alpha = 1) 
plt.scatter(*prox.T, s=1, c='red', alpha=0.5) 
plt.scatter(*loc.T, s=1, c='blue',  alpha=0.5) 
plt.xlabel('x[m]')
plt.ylabel('y[m]')
plt.xticks(np.arange(-40, 130, 40))
plt.yticks(np.arange(-180, 30, 40))
plt.savefig('C:/GNG2/mapdata/result/allroute.png')
plt.savefig('C:/GNG2/mapdata/result/allroute.eps')
plt.show()
```

```python
trave = pd.DataFrame(trave, columns = ['x', 'y'])
trave.to_csv('C:/GNG2/mapdata/merge/ku2022_trave.csv')

prox = pd.DataFrame(prox, columns = ['x', 'y'])
prox.to_csv('C:/GNG2/mapdata/merge/ku2022_prox.csv')

loc = pd.DataFrame(loc, columns = ['x', 'y'])
loc.to_csv('C:/GNG2/mapdata/merge/ku2022_loc.csv')
```
